```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
size =800
```

```python
my_array =np.array([[1,0]*size,[0,1]*size])
```

```python
plt.figure(figsize=(8,8))
plt.imshow(my_array,cmap='gray')
```

```python
x=plt.imread('yy.png')
```

```python
plt.imshow(x)
```

```python
plt.imshow(x[:,:,1],aspect='equal',cmap='gray',origin='upper',resample=False)
```
